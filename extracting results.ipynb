```python
import os
import cv2

def read_images_from_folder(folder_path):
    """Reads all images from a given folder.

    Args:
        folder_path (str): The path to the folder containing the images.

    Returns:
        list: A list of image data (NumPy arrays).
    """

    images = []
    files=os.listdir(folder_path)
    files.sort()
    for filename in files:
        # Adjust file extensions as needed
        image_path = os.path.join(folder_path, filename)
        image = cv2.imread(image_path)
        image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        image = image.astype(np.float32)
        image = cv2.resize(image, (224, 224))

        image = image / 255.0
        image= np.where(image > 0.5, 1, 0).astype(np.uint8)
        images.append(image)
    return images
```

```python
import numpy as np

def calculate_iou(mask1, mask2):
    """Calculates the Intersection over Union (IoU) of two masks.

    Args:
        mask1 (numpy.ndarray): The first mask.
        mask2 (numpy.ndarray): The second mask.

    Returns:
        float: The IoU value.
    """
    #mask1 = mask1.astype(bool)
    #mask2 = mask2.astype(bool)

    intersection = np.logical_and(mask1, mask2).sum()
    union = np.logical_or(mask1, mask2).sum()

    if union == 0:
        return 0.0

    iou = intersection / union
    return iou

def calculate_miou(masks1, masks2):
    """Calculates the mean Intersection over Union (mIoU) of two sets of masks.

    Args:
        masks1 (list): A list of masks.
        masks2 (list): A list of masks.

    Returns:
        float: The mIoU value.
    """

    miou = []
    for mask1, mask2 in zip(masks1, masks2):
        miou.append(calculate_iou(mask1, mask2))

    return np.mean(miou)

def calculate_iou_only(masks1, masks2):
    """Calculates the mean Intersection over Union (mIoU) of two sets of masks.

    Args:
        masks1 (list): A list of masks.
        masks2 (list): A list of masks.

    Returns:
        float: The mIoU value.
    """

    miou = []
    for mask1, mask2 in zip(masks1, masks2):
        miou.append(calculate_iou(mask1, mask2))

    return miou
```
